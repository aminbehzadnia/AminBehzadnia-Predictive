# **Random** **Forests**

# **Import Necessary Libraries python**


```python
# Machine learning tools
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
dataRF = Data
```

# ***Data Exploration***

```python
# Check for missing values
miss_dataRF = dataRF.isnull().sum()
print(f"the missing value\n {miss_dataRF}")

# Get information about the dataset
dataRF.info()

# Statistical summary
dataRF.describe()

Class = dataRF['class'].value_counts()
print(Class)
```

# **Data Preprocessing**

# **encoding categorical variables**

```python
le = LabelEncoder()
encoded_dataRF = dataRF.apply(le.fit_transform)
print(encoded_dataRF)
```

# **Split the Data into Training and Testing Sets**

```python
# Separate features and target variable
XRF = encoded_dataRF.drop('class', axis=1)
yRF = encoded_dataRF['class']

# Split the data into training and testing sets (80-20 split)
X_trainRF, X_testRF, y_trainRF, y_testRF = train_test_split(XRF, yRF, test_size=0.2, random_state=42, stratify=y)

# Check the shape of the datasets to confirm the split
X_trainRF.shape, X_testRF.shape, y_trainRF.shape, y_testRF.shape
```

# **Train the Random Forest Model**


```python
# Train the Random Forest Model
rf_model = RandomForestClassifier(n_estimators=100,random_state=42)
rf_model.fit(X_trainRF, y_trainRF)
```

# **Make Predictions**

```python
# Make predictions on the test set
y_predRF = rf_model.predict(X_testRF)
```

# **Evaluate the Model**

```python
# Evaluate the model
accuracyRF = accuracy_score(y_testRF, y_predRF)
print(f"AccuracyRF: {accuracyRF * 100:.2f}%")

precisionRF = precision_score(y_testRF, y_predRF, pos_label=1)
print(f"PrecisionRF: {precisionRF}")

recallRF = recall_score(y_testRF, y_predRF, pos_label=1)
print(f"RecallRF: {recallRF}")

f1RF = f1_score(y_testRF, y_predRF, pos_label=1)
print(f"F1 ScoreRF: {f1RF}")

reportRF = classification_report(y_testRF, y_predRF)
print("Classification ReportRF:\n", reportRF)

# Collecting confusion matrix
conf_matrixRF = confusion_matrix(y_testRF, y_predRF)
print("Confusion MatrixRF:\n", conf_matrixRF)
```

# **Visualization of the Confusion Matrix**

```python
# Plot the confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrixRF, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Edible', 'Poisonous'],
            yticklabels=['Edible', 'Poisonous'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()
```

# **Feature Importance**

```python
# Extract feature importances
feature_importances = rf_model.feature_importances_

# Create a DataFrame
fi_RF = pd.DataFrame({'Feature': XRF.columns, 'Importance': feature_importances})

# Sort the DataFrame
fi_RF = fi_RF.sort_values('Importance', ascending=False)

# Plot the feature importances
plt.figure(figsize=(10,6))
sns.barplot(x='Importance', y='Feature', data=fi_RF)
plt.title('Feature Importances Random Forest')
plt.show()
```
